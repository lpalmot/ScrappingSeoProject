```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import numpy as np
import re
petitdej = urlopen("http://www.aucalais.fr/breakfast.html")
lesmenus = urlopen("http://www.aucalais.fr/menu.html")
lacarte = urlopen("http://www.aucalais.fr/carte.html")
lesmoules =  urlopen("http://www.aucalais.fr/moules.html")
pizzas =  urlopen("http://www.aucalais.fr/pizza.html")
```

```python
souptitdej = BeautifulSoup(petitdej,'html.parser')
soupmenus = BeautifulSoup(lesmenus,'html.parser')
souplacarte = BeautifulSoup(lacarte,'html.parser')
souplesmoules = BeautifulSoup(lesmoules,'html.parser')
soupizza = BeautifulSoup(pizzas,'html.parser')
listsoup=[souptitdej,soupmenus,souplacarte,souplesmoules,soupizza]
```

```python
dico={}
for iteration in souptitdej.find_all('u')[0:-1]:
    productname = iteration.text
    productcomments = []
    for j in iteration.find_all_next("td")[0:-2]:
        try:
            productcomments.append(j.font.text)
        except AttributeError:
            pass
    dico[productname]=productcomments
print(dico)
```

```python
for i in dico['English Breakfast']:
    tempo = re.sub(r"[[\n]|\r|\xa0|\x80|n1]", '', i)
    print(tempo)
```
