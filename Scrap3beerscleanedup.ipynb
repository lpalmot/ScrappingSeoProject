```python
#import + déclaration des variables pour les plats
from bs4 import BeautifulSoup
from urllib.request import urlopen
import numpy as np
Apero_entree = urlopen("https://www.3brasseurs.com/aperitifs-gourmands-et-entrees-1")
soupApero= BeautifulSoup(Apero_entree,'html.parser')
Plats_bistro = urlopen("https://www.3brasseurs.com/plats-bistrot-4")
soupPlat_bistro = BeautifulSoup(Plats_bistro,'html.parser')
Flammekueches= urlopen("https://www.3brasseurs.com/flammekueches-3")
soup_flamme= BeautifulSoup(Flammekueches,'html.parser')
viandes_poissons = urlopen("https://www.3brasseurs.com/viandes-et-poissons-5")
soupvi_po = BeautifulSoup(viandes_poissons, 'html.parser')
incontournables= urlopen("https://www.3brasseurs.com/incontournables-6")
soup_inc = BeautifulSoup(incontournables, 'html.parser')
hamburgers= urlopen("https://www.3brasseurs.com/hambourgeois-7")
soup_hamb = BeautifulSoup(hamburgers, 'html.parser')
salades = urlopen("https://www.3brasseurs.com/salades-2")
soup_salade=BeautifulSoup(salades, 'html.parser')
choucroute = urlopen("https://www.3brasseurs.com/choucroutes-8")
soup_choucroute = BeautifulSoup(choucroute, 'html.parser')
desserts = urlopen("https://www.3brasseurs.com/desserts-et-flammekuechettes-9")
soup_desserts = BeautifulSoup(desserts, 'html.parser')
liste_plats = [soupApero,soupPlat_bistro,soup_flamme,soupvi_po,soup_inc,soup_hamb,soup_salade,soup_choucroute,soup_desserts]
```

```python
#déclaration des variables pour les bieres
beer1 = urlopen("https://www.3brasseurs.com/biere-blanche-1-brassee-sur-place")
soupbeer1 = BeautifulSoup(beer1, 'html.parser')
beer2 = urlopen("https://www.3brasseurs.com/biere-blonde-2-brassee-sur-place")
soupbeer2 = BeautifulSoup(beer2, 'html.parser')
beer3 = urlopen("https://www.3brasseurs.com/biere-ambree-3-brassee-sur-place")
soupbeer3 = BeautifulSoup(beer3, 'html.parser')
beer4 = urlopen("https://www.3brasseurs.com/biere-brune-4-brassee-sur-place")
soupbeer4 = BeautifulSoup(beer4, 'html.parser')
beers = [soupbeer1,soupbeer2,soupbeer3,soupbeer4]
```

```python
#fonction récupération d'un dictionnaire contenant tous les plats
def scrap(page):  
    varlist=[]
    varcomments=[]
    difference="-1"
    for iteration in page.find_all('p',{"class":"pubdish-dishname"}):
        varlist.extend(iteration.findAll(text=True))
        tempo = iteration.find_next('p',{"class":"pubdish-dishdesc"})
        if tempo != difference:
            try:
                varcomments.append(tempo.text)
                difference=tempo
            except AttributeError:
                varcomments.append("NA")
        else:
            varcomments.append("NA")

    dicovar=dict(zip(varlist, varcomments))
    return(dicovar)
```

```python
#récuperation d'un dictionnaire contenant les infos des bières
def scrap_beers(beer_list):
    dicovar={}
    for iteration in beer_list:
        beercomments=[]
        beername = iteration.find('p', {"class":"title-01 title-01--black"}).text
        beercom = iteration.find_all('p', {"class":"organoleptic-content"})
        for i in beercom:
            beercomments.extend(i.find_all(text=True))
        dicovar[beername]=beercomments
    return(dicovar)
```

```python
#appel des fonctions pour récupérer les infos completes
full_dico_bouffe={}
for iteration in liste_plats:
    tempo = scrap(iteration)
    full_dico_bouffe.update(tempo)
full_dico_beers =scrap_beers(beers)
```

```python
#Précréation du dataframe
import pandas as pd
df= pd.DataFrame(columns=["site","Produit","Type","Prix","Commentaires"])
```

```python
#transformation du 
listproducts = list(full_dico_bouffe.keys())
listproducts.extend(list(full_dico_beers.keys()))
listcomments= list(full_dico_bouffe.values())
listcomments.extend(list(full_dico_beers.values()))
```

```python
site = ["3brasseurs"]*90
Type = ["Plat"]*86
Boisson = ["Boisson"]*4
Type.extend(Boisson)
```

```python
df.Produit=listproducts
df.Commentaires=listcomments
df.site = site
df.Type = Type
df.replace("NA", np.nan )
```
